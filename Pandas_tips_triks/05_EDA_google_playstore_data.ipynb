```python
# Convert the whole size column into dytes

# let's define a function

def convert_size(size):
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', "")) * 1024
        elif 'M' in size:
            return float(size.replace('M', "")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
df['Size'].value_counts()
```

```python
# let's apply this function
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size']
```

```python
# rename the column
df.rename(columns={'Size': 'Size_in_bytes'}, inplace=True)
```

```python
df.head(3)
```

```python
df['Size_in_MB'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))
```

```python
df['Size_in_KB'] = df['Size_in_bytes'].apply(lambda x: x/1024)
```

## Install Columns:

1. Remove + sign
2. Remove ,
3. convert the column into integer

```python
# Remove the '+' from the Installs column

df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', "") if '+' in str(x) else x)
```

```python
# remove the ',' from the Installs column

df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', "") if ',' in str(x) else x)
```

```python
# Convert the column into integer

df['Installs'] = df['Installs'].apply(lambda x: int(x))
```

```python
df['Installs'].value_counts()
```

## Price Column

```python
df['Price'].value_counts()
```

```python
# Let's see how many values have $ sign

df['Price'].loc[df['Price'].str.contains('\$')].value_counts().sum()
```

```python
# The counts the number of values in the `Price` column which contains 0 but does not contain $ sign

df['Price'].loc[df['Price'].str.contains('$') & (~df['Price'].str.contains('\$'))].value_counts().sum()
```

```python
df['Price'] = df['Price'].apply(lambda x: x.replace('$', '') if '$' in str(x) else x)
```

```python
df['Price'].value_counts()
```

```python
# Now We can convert this into a numeric value

df['Price'] = df['Price'].apply(lambda x: float(x))

df['Price'].value_counts()
```

```python
df.describe()
```

```python
# using f string print min, max & average prices of the apps
print(f"The minimum price of the apps is {df['Price'].min()}")
print(f"The maximum price of the apps is {df['Price'].max()}")
print(f"The average price of the apps is {df['Price'].mean()}")
```

## Missing values inside the dataframe

```python
# find the missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# find the missing values percentage in the data
round(df.isnull().sum()/len(df)*100, 2).sort_values(ascending=False)
```

```python
# find the number of missing values in each column
df.isnull().sum().sum()
```

```python
# plot missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# make figure size
plt.figure(figsize=(10, 6))
# plot the null values by their percentage in each column
missing_percentage = round(df.isnull().sum()/len(df)*100, 2)
missing_percentage.plot(kind='bar')

# add the labels
plt.xlabel('Features')
plt.ylabel('Percentage of missing values')
plt.title('Percentage of missing values in each column')
plt.show() 
```

```python
missing_percentage[missing_percentage < 1].plot(kind='bar')
```
